```python
%matplotlib inline
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#根据经验，剔除原始数据中的异常值，使得各维度的数据尽可能地符合正态分布
in_data = pd.read_excel('hw_normal.xlsx')
in_data.head()
```

```python
#预测2G北美市场规模-创建数据集
twogna_train_x = []
twogna_train_y = []

for _ in range(0,88):
    if in_data['2G北美'][_] != 0:
        twogna_train_y.append(in_data['2G北美'][_])
        twogna_train_x.append(_)
twogna_train_x
twogna_train_y

twogna_pre_x = []
for _ in range(0,88):
    if in_data['2G北美'][_] == 0:
        twogna_pre_x.append(_)
twogna_pre_x[-6:-1]
```

```python
#将各区域的 3G(0~43)、4G(40~83)数据的第2～12年的数据 按照权重比为 4:6做加权处理，生成新的数据维度
quan_na = []
quan_emea = []
quan_yt = []
quan_cala = []

for i in range(0,44):
    indivi_quan_na = in_data['3G北美'][i] * 0.4 + in_data['4G北美'][i+40] * 0.6
    quan_na.append(indivi_quan_na)
    indivi_quan_emea = in_data['3G EMEA'][i] * 0.4 + in_data['4G EMEA'][i+40] * 0.6
    quan_emea.append(indivi_quan_emea)
    indivi_quan_yt = in_data['3G亚太'][i] * 0.4 + in_data['4G亚太'][i+40] * 0.6
    quan_yt.append(indivi_quan_yt)
    indivi_quan_cala = in_data['3G CALA'][i] * 0.4 + in_data['4G CALA'][i+40] * 0.6
    quan_cala.append(indivi_quan_cala)
    
df_quan = pd.DataFrame(
    {
        'quan_na':quan_na,
        'quan_emea':quan_emea,
        'quan_yt':quan_yt,
        'quan_cala':quan_cala
    }
)

df_quan
```

```python
#预测四个地区 5g 规模 创建数据集
fivegna_train_x = []
fivegna_train_y = []
fivegemea_train_x = []
fivegemea_train_y = []
fivegyt_train_x = []
fivegyt_train_y = []
fivegcala_train_x = []
fivegcala_train_y = []

for i in range(0,44):
    fivegna_train_x.append(i)
    fivegna_train_y.append(df_quan['quan_na'][i])
    fivegemea_train_x.append(i)
    fivegemea_train_y.append(df_quan['quan_emea'][i])
    fivegyt_train_x.append(i)
    fivegyt_train_y.append(df_quan['quan_yt'][i])
    fivegcala_train_x.append(i)
    fivegcala_train_y.append(df_quan['quan_cala'][i])
    
fivegna_train_x = np.array(fivegna_train_x).reshape(-1,1)
fivegemea_train_x = np.array(fivegemea_train_x).reshape(-1,1)
fivegyt_train_x = np.array(fivegyt_train_x).reshape(-1,1)
fivegcala_train_x = np.array(fivegcala_train_x).reshape(-1,1)

fivegna_train_y = np.array(fivegna_train_y)
fivegemea_train_y = np.array(fivegemea_train_y)
fivegyt_train_y = np.array(fivegyt_train_y)
fivegcala_train_y = np.array(fivegcala_train_y)

fivegna_pre_x = [-4,-3,-2,-1]
fivegemea_pre_x = [-4,-3,-2,-1]
fivegyt_pre_x = [-4,-3,-2,-1]
fivegcala_pre_x = [-4,-3,-2,-1]
```

```python
#预测北美地区 5g规模 -- 创建随机森林模型
Xtrain,Xtest,Ytrain,Ytest = train_test_split(fivegna_train_x,fivegna_train_y,test_size=0.3)

rfr_5gna = RandomForestRegressor(random_state=15,n_estimators=100)
rfr_5gna = rfr_5gna.fit(Xtrain,Ytrain)

score_rfr_5gna = rfr_5gna.score(Xtest,Ytest)
score_rfr_5gna
```

```python
#预测EMEA地区 5g规模 -- 创建随机森林模型
Xtrain,Xtest,Ytrain,Ytest = train_test_split(fivegemea_train_x,fivegemea_train_y,test_size=0.3)

rfr_5gemea = RandomForestRegressor(random_state=15,n_estimators=100)
rfr_5gemea = rfr_5gemea.fit(Xtrain,Ytrain)

score_rfr_5gemea = rfr_5gemea.score(Xtest,Ytest)
score_rfr_5gemea
```

```python
#预测亚太地区 5g规模 -- 创建随机森林模型
Xtrain,Xtest,Ytrain,Ytest = train_test_split(fivegyt_train_x,fivegyt_train_y,test_size=0.3)

rfr_5gyt = RandomForestRegressor(random_state=20,n_estimators=100)
rfr_5gyt = rfr_5gyt.fit(Xtrain,Ytrain)

score_rfr_5gyt = rfr_5gyt.score(Xtest,Ytest)
score_rfr_5gyt
```

```python
#预测cala地区 5g规模 -- 创建随机森林模型
Xtrain,Xtest,Ytrain,Ytest = train_test_split(fivegcala_train_x,fivegcala_train_y,test_size=0.3)

rfr_5gcala = RandomForestRegressor(random_state=10,n_estimators=100)
rfr_5gcala = rfr_5gcala.fit(Xtrain,Ytrain)

score_rfr_5gcala = rfr_5gcala.score(Xtest,Ytest)
score_rfr_5gcala
```

```python
#预测4个地区的 5g规模
fivegna_pre_x = np.array(fivegna_pre_x).reshape(-1,1)
fivegemea_pre_x = np.array(fivegemea_pre_x).reshape(-1,1)
fivegyt_pre_x = np.array(fivegyt_pre_x).reshape(-1,1)
fivegcala_pre_x = np.array(fivegcala_pre_x).reshape(-1,1)

fivegna_pre_y = rfr_5gna.predict(fivegna_pre_x)
fivegemea_pre_y = rfr_5gemea.predict(fivegemea_pre_x)
fivegyt_pre_y = rfr_5gyt.predict(fivegyt_pre_x)
fivegcala_pre_y = rfr_5gcala.predict(fivegcala_pre_x)
```
